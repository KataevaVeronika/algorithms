```python
import random
import datetime
import itertools
import numpy as np
import networkx as nx

from itertools import combinations
```

```python
vertices = list(range(1, 101))
edges = random.sample(list(combinations(vertices, 2)), 500)
weights = np.random.randint(1, 100, size=500)
weighted_edges = [(*edge, weight) for edge, weight in zip(edges, weights)]
```

```python
G = nx.Graph()
G.add_weighted_edges_from(weighted_edges)
print(f'Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}')
```

```python
nx.draw(G)
```

```python
def calculate_time(method):
    times = []
    for _ in range(10):
        timestamp_1 = datetime.datetime.now()
        nx.shortest_path(G, source=1, target=100, method=method)
        timestamp_2 = datetime.datetime.now()
        times.append((timestamp_2 - timestamp_1).total_seconds())
    return np.average(times)
```

```python
print(f'Average time of Dijkstra algorithm: {calculate_time("dijkstra"):.7f} s')
```

```python
print(f'Average time of Bellman-Ford algorithm: {calculate_time("bellman-ford"):.7f} s')
```

```python
G = nx.grid_graph((10, 20))
```

```python
nodes = list(G.nodes())
indices_to_remove = np.random.choice(200, size=40, replace=False)
nodes_to_remove = [nodes[n] for n in indices_to_remove]
```

```python
G.remove_nodes_from(nodes_to_remove)
G.number_of_nodes()
```

```python
positions = {}
for tuple_ in G.nodes():
    positions.update({tuple_: list(tuple_)})
```

```python
def find_shortest_path(G):
    nodes = list(G.nodes())
    indices_for_path = np.random.choice(160, size=2, replace=False)
    nodes_for_path = [nodes[n] for n in indices_for_path]
    
    shortest_path = nx.astar_path(G, nodes_for_path[0], nodes_for_path[1])
    
    path_positions = {}
    for tuple_ in shortest_path:
        path_positions.update({tuple_: list(tuple_)})
    
    nx.draw(G, pos=positions, node_size=200)
    nx.draw_networkx_nodes(G, pos=path_positions, nodelist=shortest_path, node_color='red')
```
